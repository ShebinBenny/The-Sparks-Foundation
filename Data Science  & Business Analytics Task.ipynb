# Exploratory Data Analysis - Terrorism

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

reading the csv file

```python
df=pd.read_csv('globalterrorismdb_0718dist.csv',encoding='latin1',low_memory=False) 
```

```python
df
```

```python
#Statistical Description of the dataset
df.describe()
```

```python
#To find the dimension of our dataset
df.shape
```

```python
#To know the various attributes in our dataset
df.columns
```

```python
#renaming columns for understanding and easier reuse
df.rename(columns=({'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'}),inplace=True)
```

```python
# Choosing only the important columns
df=df[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
df.head()
```

```python
#To display the null values for each attribute in our dataset
df.isnull().sum()
```

```python
#To find what type of datatype each column have
df.info()
```

Visualizations

```python
# To display increasing number of terror activities per year
x_year = df['Year'].unique()
y_count_years = df['Year'].value_counts().sort_index()
plt.figure(figsize = (20,10))
sns.barplot(x = x_year,
           y = y_count_years)
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack Vs Years')
plt.show()     
```

```python
#Bar Plot - Top 15 countries that are affected
plt.figure(figsize=(10,6))
sns.barplot(df['Country'].value_counts()[:15].index,df['Country'].value_counts()[:15].values,palette='rocket')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()    
```

```python
#Number of people killed in a terror attack
killData = df.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))
```

```python
#Country wise Death
countryData = df.loc[:,'Country']
countryKillData = pd.concat([countryData, killData], axis=1)
countryKillData
```

```python
#Country wise Death
countryKillFormatData = countryKillData.pivot_table(columns='Country', values='Killed', aggfunc='sum')
countryKillFormatData
```

```python
#Bar Plot - Weapon types used in attacks
plt.figure(figsize=(15, 9))
sns.barplot(
    df["Weapon_type"].value_counts()[:10].index,
    df["Weapon_type"].value_counts()[:10].values,
    palette=("rocket"),
)
plt.title("Weapon types used in attacks")
plt.xlabel("Weapon of choice")
plt.ylabel("Counts")
plt.xticks(rotation=30)
plt.show(); 
```

```python
plt.figure(figsize=(15, 9))
sns.barplot(
    df["Target_type"].value_counts()[:10].index,
    df["Target_type"].value_counts()[:10].values,
    palette=("Blues_d"),
)
plt.title("Target types affected by attacks")
plt.xlabel("Target")
plt.ylabel("Counts")
plt.xticks(rotation=90)
plt.show();
```
